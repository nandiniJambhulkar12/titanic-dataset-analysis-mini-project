```python
# kde:kernal density estimation
```

```python
df[df["Survived"]==0]["Age"]
```

```python
sns.kdeplot(df[df["Survived"]==1]["Age"],color="green",label="Survived")
sns.kdeplot(df[df["Survived"]==0]["Age"],color="red",label="NotSurvived")
plt.legend()
plt.show()
```

```python
# survival vs fare
```

```python
sns.kdeplot(df[df["Survived"]==1]["Fare"],color="green",label="Survived")
sns.kdeplot(df[df["Survived"]==0]["Fare"],color="red",label="NotSurvived")
plt.legend()
plt.show()
```
